```python
img2=cv2.resize(img2,(600,300))
img1=cv2.resize(img1,(600,300))
```

```python
plt.imshow(img2[45:165,155:455])#Slicing object
```

```python
#Displaying resized img2
cv2.imshow('Take-Off',img2)
cv2.waitKey()
```

```python
#Blurring the flight 
flight_blur=img2.copy()
roi=flight_blur[45:165,155:455]
flight_blur[45:165,155:455]=cv2.blur(roi,(50,50))
cv2.imshow('Take-Off',flight_blur)
cv2.waitKey()
```

```python
#Detecting Edges
cv2.imshow('Cars',cv2.Canny(img1,250,500))
cv2.imshow('Take-Off',cv2.Canny(img2,250,500))
cv2.waitKey()
```

# Working with Video Processing

```python
video=cv2.VideoCapture(0)
while True:
    data,frame=video.read()
    cv2.imshow("My Video",cv2.blur(frame,(100,100)))
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
